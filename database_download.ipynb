## Data download
Download the database of the spoken numbers. The database contains 20,000 samples, which can be used for training/validation/testing.

```python
# a function for downloading a file into a chosen location
def download_url(url, save_path):
    with urllib.request.urlopen(url) as dl_file:
        with open(save_path, 'wb') as out_file:
            out_file.write(dl_file.read())
```
